```python
from pyspark.context import SparkContext
from pyspark.sql.session import SparkSession
from pyspark.sql.functions import year
from pyspark.streaming import StreamingContext
from pyspark.sql.types import StructType, StructField, StringType, DoubleType, IntegerType

sc = SparkContext('local')
spark = SparkSession(sc)
ssc = StreamingContext(sc, 10)

# Define the schema for the CSV file
schema = StructType([
    StructField("Reported_date", StringType(), True),
    StructField("Reported_time", StringType(), True),
    StructField("Description", StringType(), True),
    StructField("Address", StringType(), True),
    StructField("City", StringType(), True),
    StructField("Zip_code", StringType(), True),
    StructField("State", StringType(), True),
    StructField("Country", StringType(), True),
    StructField("Reported_year", IntegerType(), True),
    StructField("Reported_month", IntegerType(), True),
    StructField("Reported_day", IntegerType(), True)
])
```

```python
# Define the path to the input CSV file
input_path = "file:/content/drive/MyDrive/DBMS_project/Crime-Rate-Analysis-master/Crime-Rate-Analysis-master/Source Code/1-Datasets/KCPD_Crime_Data/data_un/new"

# Define the streaming DataFrame that reads from the CSV file
stream_df = spark.readStream.schema(schema).csv(input_path)

# Filter the DataFrame to include only rows with a Reported_year of 2013
year_filtered_df = stream_df.filter(year("Reported_date") == 2013)

# Define the output path for the CSV files
output_path = "/content/drive/MyDrive/DBMS_project/Crime-Rate-Analysis-master/Crime-Rate-Analysis-master/Source Code/1-Datasets/KCPD_Crime_Data/data_use/"

# Write the filtered DataFrame to the output CSV files, divided by year
stream_query = year_filtered_df.writeStream.format("csv").option("path", output_path).option("checkpointLocation", "/tmp/checkpoint").partitionBy("Reported_year").start()

# Wait for the stream to finish
stream_query.awaitTermination()
```
